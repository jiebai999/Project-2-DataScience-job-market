```python
# Dependencies and Setup
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
```

```python
# Set Executable Path & Initialize Chrome Browser
executable_path = {"executable_path": "chromedriver.exe"}
browser = Browser("chrome", **executable_path)
```

### Indeed Data Scientist Jobs

```python
# Visit the Indeed Website
url = "https://www.indeed.com/jobs?q=data+scientist&l="
browser.visit(url)
```

```python
# Create Beautifulsoup object; parse with html.parser
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
#scraping the job posting blocks
latest_job_title = soup.find_all('div', class_='jobsearch-SerpJobCard')
latest_job_title 
```

```python
latest_job_title[0].find('span',class_="company").text.replace("\n","")
```

```python
job_title = []
job_location = []
job_skills = []
job_salary = []
job_company = []

scraping = True
while scraping == True:
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    jobs = soup.find_all('div', class_='jobsearch-SerpJobCard')
    
    for job in jobs:
        jobs_title = job.find('a',class_="jobtitle")
        title = jobs_title["title"]
        print(title)        
        job_title.append(title)
        
        locations = job.find('div',class_='recJobLoc')
        location = locations["data-rc-loc"]
        print(location)        
        job_location.append(location)
        
        jobs_company = job.find('span',class_="company").text.replace("\n","")
        print(jobs_company)        
        job_company.append(jobs_company)
                
    try:
        browser.click_link_by_partial_text('Next')
    except:
        close_button = browser.find_by_id('popover-x')
        close_button.click()
    else:
        scraping = True
```

```python
df = pd.DataFrame({
    "Job_Title":job_title,
    "Location":job_location,
    "Company":job_company
})
df.head()
```
